```python
import seaborn as sns 
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.heatmap(df.corr(),annot=True)

# heatmap only for correation matrix 
```

```python
# jointplot

sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='resid')
```

```python
# pair plot 
# combine all interger column to shown in graph

sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
# dist plot

sns.distplot(df['tip'])
```

```python
# kde - kernael density estimation

sns.distplot(df['tip'],kde=False)
```

## Categorical plots

```python
# countplot

sns.countplot('sex',data=df)
```

```python
sns.countplot(data=df,y='smoker')
```

```python
# bar plot

sns.barplot(x='total_bill',y='sex',data=df)
```

```python
# box plot

sns.boxplot('tip','sex',data=df)
```

```python
sns.boxplot('day','tip',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='h')
```

```python
sns.boxplot(x='total_bill',y='day',hue='sex',data=df)
```

```python
# violin plot
# combination of KDE and boxplot

sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
